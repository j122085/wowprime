```python
with open("../data/CostPower.csv") as f:
    x=f.read()
```

```python
keys=x.replace("\t",",").split("\n")[0].split(",")
keys
```

```python
values=[data.split(",") for data in x.replace("\t",",").split("\n")[1:]]
values
```

```python
datalist=[]
for valueList in values:
    data={}
    for key,value in zip(keys,valueList):
        data[key]=value
        datalist.append(data)
```

```python
datalist
```

```python
dataFinal=[]
for i in datalist:
    data={}
    if i.get('主行業別',"")=='餐飲業' and i.get('年度',"")=='2017' and i.get("消費熱度計算來源","")=="電子發票":
        data['add']=i['縣市']+i['鄉鎮市區']+i['村里']
        data['costPower']=i['銷售額指標']
        dataFinal.append(data)
```
