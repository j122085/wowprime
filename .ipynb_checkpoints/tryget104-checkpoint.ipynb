```python
import requests
from bs4 import BeautifulSoup as bs
import re
URL="https://www.104.com.tw/jobs/search/?ro=0&jobcat=2006001000&order=2&asc=0&mode=s&jobsource=n104bank1"
```

```python
res=requests.get(URL+"&page=1")
```

```python
res.status_code
```

```python
pages=int(res.text.split('"totalPage":')[1].split(',"totalCount"')[0])
pages
```

```python
import time
a=set()
for page in range(1,pages+1):
    res=requests.get(URL+"&page=%s"%page)
    if res.status_code==200:
        soup=bs(res.text,'lxml')
        res.close()
        a.update(["https:"+job['href'].split("&jobsource")[0] for job in soup.select("a.js-job-link ")])
        print(URL+"&page=%s"%page)
        print(len(a))
        time.sleep(3)
    else:
        print("!!!!")
#js-job-header > div.b-float-right > label:nth-child(1) > select > option:nth-child(1)
```
